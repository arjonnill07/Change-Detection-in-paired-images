```python
# Install required libraries
!pip install ultralytics opencv-python-headless==4.8.0.74 matplotlib

# Import necessary libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Mount Google Drive to access the dataset
from google.colab import drive
drive.mount('/content/drive')

# Define paths to the dataset
train_path = "/content/drive/MyDrive/LEVIR-CD+/train"
test_path = "/content/drive/MyDrive/LEVIR-CD+/test"

# Define paths for training images and labels
train_A_path = os.path.join(train_path, "A")
train_B_path = os.path.join(train_path, "B")
train_label_path = os.path.join(train_path, "label")

test_A_path = os.path.join(test_path, "A")
test_B_path = os.path.join(test_path, "B")
test_label_path = os.path.join(test_path, "label")
```

```python

# Define YOLO dataset preparation function
def mask_to_yolo(mask):
    """
    Convert binary masks to YOLO format annotations.
    """
    height, width = mask.shape
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    annotations = []
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        x_center = (x + w / 2) / width
        y_center = (y + h / 2) / height
        w_norm = w / width
        h_norm = h / height
        annotations.append(f"0 {x_center} {y_center} {w_norm} {h_norm}")
    return annotations

def prepare_dataset(image_A_path, image_B_path, label_path, save_path):
    """
    Prepare the YOLO dataset by combining paired images and generating YOLO labels.
    """
    os.makedirs(os.path.join(save_path, "images"), exist_ok=True)
    os.makedirs(os.path.join(save_path, "labels"), exist_ok=True)
    img_names = sorted(os.listdir(image_A_path))

    for img_name in img_names:
        img_A = cv2.imread(os.path.join(image_A_path, img_name))
        img_B = cv2.imread(os.path.join(image_B_path, img_name))
        label = cv2.imread(os.path.join(label_path, img_name), cv2.IMREAD_GRAYSCALE)

        # Combine images (side by side)
        combined_image = np.concatenate((img_A, img_B), axis=1)

        # Convert binary mask to YOLO format annotations
        annotations = mask_to_yolo(label)

        # Save combined image and annotations
        cv2.imwrite(os.path.join(save_path, "images", img_name), combined_image)
        with open(os.path.join(save_path, "labels", img_name.split('.')[0] + ".txt"), 'w') as f:
            f.write("\n".join(annotations))

# Prepare the dataset for YOLO
train_save_path = "/content/yolo_train"
test_save_path = "/content/yolo_test"

prepare_dataset(train_A_path, train_B_path, train_label_path, train_save_path)
prepare_dataset(test_A_path, test_B_path, test_label_path, test_save_path)

# Visualize a sample combined image and corresponding label
sample_image_path = os.path.join(train_save_path, "images", "train_1.png")
sample_label_path = os.path.join(train_save_path, "labels", "train_1.txt")

sample_image = cv2.imread(sample_image_path)
plt.imshow(cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB))
plt.title("Sample Combined Image")
plt.show()

with open(sample_label_path, "r") as f:
    print("Sample YOLO Label:")
    print(f.read())

# Initialize YOLO model and train
model = YOLO("yolov8n.pt")  # Using YOLOv8 Nano for efficiency

model.train(
    data={
        "train": os.path.join(train_save_path, "images"),
        "val": os.path.join(test_save_path, "images"),
        "names": ["change"]
    },
    epochs=20,
    batch=16,
    imgsz=640,
    project="change_detection_project",
    name="yolo_change_detection"
)

# Evaluate the model on the test set
results = model.val(data=os.path.join(test_save_path, "images"))
print(results)

# Perform inference on test images
test_image_path = os.path.join(test_save_path, "images", "train_638.png")
results = model.predict(test_image_path, save=True)

# Visualize the predictions
predicted_image_path = os.path.join("runs", "detect", "predict", "train_638.jpg")
predicted_image = cv2.imread(predicted_image_path)
plt.imshow(cv2.cvtColor(predicted_image, cv2.COLOR_BGR2RGB))
plt.title("Predicted Changes")
plt.show()
```
